```python
# Skewness Analysis
# Select numeric columns
num_col = df.select_dtypes(include = "number").columns
# Compute skewness
# Compute skewness
skewness = df[num_col].skew().sort_values(key=abs, ascending=False)

skewness_df = pd.DataFrame({
    'Skewness': skewness
})
skewness_df
```

```python
# Display the two most skewed columns
top2_skew = skewness_df.head(2)

print(top2_skew)
```

The skewness analysis shows that VitD_levels (3.45) and Population (2.23) have the highest positive skewness among all numeric variables. A positive skew indicates that most observations are concentrated at lower values, with a long tail extending toward higher values due to a few extreme observations. In such distributions, the mean is influenced by these extreme values, whereas the median remains more representative of the central tendency. Therefore, the median is a more suitable choice for imputing missing values in positively skewed data.

```python
# Detect outliers using the Interquartile Range (IQR) method
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = data[(data[column] < lower) | (data[column] > upper)]

    print(f"\nColumn : {column}")
    print(f"Q1 : {Q1:.2f}")
    print(f"Q3 : {Q3:.2f}")
    print(f"IQR : {IQR:.2f}")
    print(f"Lower Bound : {lower:.2f}")
    print(f"Upper Bound : {upper:.2f}")
    print(f"Number of Outliers : {len(outliers)}")

# Run for two columns
detect_outliers_iqr(df, "VitD_levels")
detect_outliers_iqr(df, "Population")
```

* The VitD_levels column has 534 outliers based on the IQR method. These values fall outside the lower bound (11.60) and upper bound (24.70). Since Vitamin D levels can naturally vary among patients due to health conditions, these observations may represent genuine medical variations rather than data entry errors. Therefore, the outliers will be retained for further analysis.
* The Population column has 855 outliers according to the IQR method. These represent locations with exceptionally high population sizes. Since population values naturally vary across different geographic regions, these observations are likely to be valid rather than erroneous. Therefore, the outliers will be retained instead of removed.

```python
# Display Columns with More Than 20% Missing
# Columns with more than 20% missing values

missing_percent = (df.isnull().sum() / len(df)) * 100

high_missing = missing_percent[missing_percent > 20]

print(high_missing)
```

Four columns (Children, Age, Income, Soft_drink) contain more than 20% missing values. These columns were not imputed during Part 1 because a high percentage of missing data may require more advanced handling techniques.

```python
# Fill numeric columns having less than 20% missing values with median

missing_percent = (df.isnull().sum() / len(df)) * 100

numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

for col in numeric_cols:
    if 0 < missing_percent[col] < 20:
        df[col] = df[col].fillna(df[col].median())
```

```python
# Check remaining missing values

remaining_missing = df.isnull().sum()

print(remaining_missing[remaining_missing > 0])
```

```python
# Fill remaining numeric columns with median

df['Children'] = df['Children'].fillna(df['Children'].median())

df['Age'] = df['Age'].fillna(df['Age'].median())

df['Income'] = df['Income'].fillna(df['Income'].median())

# Fill categorical column with mode

df['Soft_drink'] = df['Soft_drink'].fillna(df['Soft_drink'].mode()[0])
```

```python
# Check for remaining missing values

print(df.isnull().sum())
```

The remaining missing values were successfully handled using appropriate imputation techniques. Median was used for numeric columns, and mode was used for the categorical column. As a result, the dataset now contains no missing values and is ready for the next phase of the project.

```python
# Compute Pearson and Spearman correlation matrices

pearson_corr = df.corr(numeric_only=True)

spearman_corr = df.corr(method='spearman', numeric_only=True)

print("Pearson Correlation Matrix")
print(pearson_corr)

print("\nSpearman Correlation Matrix")
spearman_corr
```

```python
# Compute absolute difference between Pearson and Spearman correlations

difference = (spearman_corr - pearson_corr).abs()

# Remove self-correlation
np.fill_diagonal(difference.values, 0)

difference.head()
```

```python
# Convert matrix into pairwise table

difference_pairs = (
    difference.stack()
    .reset_index()
)

difference_pairs.columns = ['Feature 1', 'Feature 2', '|Spearman - Pearson|']

# Remove duplicate pairs
difference_pairs = difference_pairs[
    difference_pairs['Feature 1'] < difference_pairs['Feature 2']
]

# Top 3 largest differences
top3 = difference_pairs.sort_values(
    by='|Spearman - Pearson|',
    ascending=False
).head(3)

print(top3)
```

```python
# Display Pearson and Spearman values for the top 3 pairs

for _, row in top3.iterrows():

    col1 = row['Feature 1']
    col2 = row['Feature 2']

    print(f"\n{col1}  <-->  {col2}")

    print("Pearson  :", round(pearson_corr.loc[col1, col2],4))
    print("Spearman :", round(spearman_corr.loc[col1, col2],4))
    print("Difference:", round(row['|Spearman - Pearson|'],4))
```

The Pearson and Spearman correlation matrices were compared to identify differences between linear and monotonic relationships. The largest difference was observed between TotalCharge and VitD_levels, where the Pearson correlation (0.7283) was much higher than the Spearman correlation (0.2303). This suggests that their apparent linear relationship is influenced by extreme values or the distribution of the data.

The relationship between Initial_days and TotalCharge showed a stronger Spearman correlation (0.8419) than Pearson (0.6424), indicating a strong monotonic relationship that is not perfectly linear.

For Lat and Lng, both correlation values were relatively weak, and the difference suggests that there is no strong monotonic relationship despite some variation in the linear measure.
